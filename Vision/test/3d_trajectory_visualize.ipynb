```python

# x, y 값의 범위를 지정하고 그에 해당하는 그리드를 생성
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# z 값 계산 (포물선 경로) -->> ???  포물면 
z = x**2 + y**2

# 그림 객체 및 3D 축 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D 서피스 플롯 생성
ax.plot_surface(x, y, z, cmap='viridis')

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Parabolic Surface: z = x^2 + y^2')

# 시각화
plt.show()
```

### 포물선 궤적 3d plot 

```python
# 매개 변수 및 값의 범위 설정
m = 1
a = -1
b = 1
c = 0
x = np.linspace(-5, 5, 100)
y = m * x

# z 값 계산
z = -(y**2 + a*y + c) / b

# 그림 객체 및 3D 축 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D 선 플롯 생성
ax.plot(x, y, z, label='Parabolic Curve')
ax.legend()

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Parabolic Curve') 

# 시각화
plt.show()
```

```python
def plot_parabola(m=1, a=-1, b=1, c=0):
    if m != 0: 
        x = np.linspace(-5, 5, 100)
        y = m * x
    else: 
        x = np.linspace(1, 1, 100)
        y = np.linspace(-5, 5, 100)
    z = -(y**2 + a*y + c) / b
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z, label='Parabolic Curve')
    
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
    ax.set_title('3D Parabolic Curve')
    ax.legend()
    
    plt.show()

# ipywidgets를 사용하여 인터렉티브 슬라이더 생성
interact(plot_parabola, m=(-5, 5, 0.1), a=(-5, 5, 0.1), b=(-5, 5, 0.1), c=(-5, 5, 0.1));
```

```python
def plot_parabolas(m1=1, a1=-1, b1=1, c1=0, m2=-1, a2=1, b2=-1, c2=0):
    x = np.linspace(-5, 5, 100)
    
    # 첫 번째 포물선 계산
    y1 = m1 * x
    z1 = -(y1**2 + a1*y1 + c1) / b1
    
    # 두 번째 포물선 계산
    y2 = m2 * x
    z2 = -(y2**2 + a2*y2 + c2) / b2
    
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y1, z1, label='Parabolic Curve 1', color='r')
    ax.plot(x, y2, z2, label='Parabolic Curve 2', color='b')
    
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')
    ax.set_title('3D Parabolic Curves')
    ax.legend()
    
    plt.show()

# ipywidgets를 사용하여 인터렉티브 슬라이더 생성
interact(plot_parabolas, 
         m1=(-5, 5, 0.1), a1=(-5, 5, 0.1), b1=(-5, 5, 0.1), c1=(-5, 5, 0.1),
         m2=(-5, 5, 0.1), a2=(-5, 5, 0.1), b2=(-5, 5, 0.1), c2=(-5, 5, 0.1));
```
